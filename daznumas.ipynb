```python
import nltk   
from urllib import urlopen
#import sys, locale
#reload(sys)
#sys.setdefaultencoding("utf-8")
import HTMLParser
import BeautifulSoup
```

```python
h = HTMLParser.HTMLParser()
```

```python
corpus = []

for i in range(1,5):
    url = "http://eur-lex.europa.eu/legal-content/LT/TXT/HTML/?uri=OJ:L:2015:00" + str(i) + ":FULL&from=LT"    
    raw = urlopen(url).read()
    
    
    corpus.append((BeautifulSoup.BeautifulSoup(h.unescape(raw))).findAll(text=True))
    #print (BeautifulSoup.get(h.unescape(raw[0:200])))
```
